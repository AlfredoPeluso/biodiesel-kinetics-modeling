```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Imposta stile grafico professionale
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
```

```python
# Percorso del file
file_path = os.path.join('..', 'data', 'experimental_data.csv')

# Carica CSV
df = pd.read_csv(file_path)

# Visualizza le prime righe
print("Prime 5 righe:")
print(df.head())

print("\n\nStatistiche descrittive:")
print(df.describe())
```

```python
# Separa i dati per temperatura
data_50 = df[df['temperature_c'] == 50]
data_60 = df[df['temperature_c'] == 60]

# Crea il grafico
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(data_50['time_min'], data_50['conversion'] * 100, 
        'o-', color='blue', linewidth=2, markersize=8, label='50°C')
ax.plot(data_60['time_min'], data_60['conversion'] * 100, 
        's-', color='red', linewidth=2, markersize=8, label='60°C')

ax.set_xlabel('Tempo (min)', fontsize=12)
ax.set_ylabel('Conversione esteri metilici (%)', fontsize=12)
ax.set_title('Transesterificazione dell\'olio di soia\nDati sperimentali (Freedman et al., 1986)', 
             fontsize=14, fontweight='bold')
ax.legend(loc='lower right', fontsize=11)
ax.grid(True, alpha=0.3)

# Aggiungi una griglia più fine
ax.minorticks_on()
ax.grid(True, which='minor', alpha=0.2)

plt.tight_layout()
plt.savefig(os.path.join('..', 'output', 'experimental_data.png'), dpi=300, bbox_inches='tight')
plt.show()

print("Grafico salvato in output/experimental_data.png")
```

```python
# Calcola la velocità iniziale (primi 5 minuti)
for temp in [50, 60]:
    data_temp = df[df['temperature_c'] == temp]
    initial_rate = data_temp[data_temp['time_min'] == 5]['conversion'].values[0] / 5
    print(f"Velocità iniziale a {temp}°C: {initial_rate:.3f} %/min")
```
